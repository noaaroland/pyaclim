```python
def add_lat(s):
    return latitudes[s]
def add_lon(s):
    return longitudes[s]
```

```python
dsg_data_array = xr.open_dataset("ACLIMregion_B10K-H16_CMIP5_CESM_BIO_rcp85.nc")
```

```python
df = dsg_data_array.to_dataframe()
df
```

```python
df = df.reset_index()
df['region'] = df['region'].astype(int)
df['latitude'] = df['region'].apply(add_lat)
df['longitude'] = df['region'].apply(add_lon)
df
```

```python
df['region'] = df['region'].astype(str)
df = df.set_index(["ocean_time", "region"])
df
```

```python
aug_xr = df.to_xarray()
aug_xr.attrs =dsg_data_array.attrs
for var_name, values in dsg_data_array.items():
    var = getattr(dsg_data_array, var_name)
    aug_var = getattr(aug_xr, var_name)
    aug_var.attrs = var.attrs
aug_xr.to_netcdf("regions_lat_lon.nc")
```

```python
# alternatively write the data frame to a  OrthogonalMultidimensionalTimeseries (PyFerret can't read as DSG)
df = df.reset_index() # has to be "flat"
axes = {"t": "ocean_time", "x": "longitude", "y": "latitude", "z":None, "station": "region"}
attributes = {"global":dict(dsg_data_array.attrs)}
for var_name, values in dsg_data_array.items():
    var = getattr(dsg_data_array, var_name)
    attributes[var_name] = dict(var.attrs)
df = downcast_dataframe(df)  # safely cast depth np.int64 to np.int32
dsg = OrthogonalMultidimensionalTimeseries.from_dataframe(
    df, output="regions_as_om.nc", attributes=attributes, axes=axes,
)
```
